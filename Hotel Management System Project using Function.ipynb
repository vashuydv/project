```python
#Hotel Management System Project using Function
rooms = {}

def add_room():
    room_number = input("Enter room number: ")
    capacity = int(input("Enter room capacity: "))
    status = "Available"
    room = {"capacity": capacity, "status": status}
    rooms[room_number] = room
    print("Room added successfully.")

def display_rooms():
    if not rooms:
        print("No rooms found.")
    else:
        print("List of rooms:")
        for room_number, room in rooms.items():
            print(f"Room Number: {room_number}")
            print(f"Capacity: {room['capacity']}")
            print(f"Status: {room['status']}")

def search_room():
    room_number = input("Enter room number to search: ")
    room = rooms.get(room_number)
    if room:
        print(f"Room Number: {room_number}")
        print(f"Capacity: {room['capacity']}")
        print(f"Status: {room['status']}")
    else:
        print("Room not found.")

def reserve_room():
    room_number = input("Enter room number to reserve: ")
    room = rooms.get(room_number)
    if room:
        if room['status'] == "Available":
            room['status'] = "Reserved"
            print("Room reserved successfully.")
        else:
            print("Room is already reserved.")
    else:
        print("Room not found.")

def checkout_room():
    room_number = input("Enter room number to checkout: ")
    room = rooms.get(room_number)
    if room:
        if room['status'] == "Reserved":
            room['status'] = "Available"
            print("Room checked out successfully.")
        else:
            print("Room is not currently reserved.")
    else:
        print("Room not found.")

def main():
    while True:
        print("\nHotel Management System")
        print("1. Add Room")
        print("2. Display Rooms")
        print("3. Search Room")
        print("4. Reserve Room")
        print("5. Checkout Room")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ")
        
        if choice == '1':
            add_room()
        elif choice == '2':
            display_rooms()
        elif choice == '3':
            search_room()
        elif choice == '4':
            reserve_room()
        elif choice == '5':
            checkout_room()
        elif choice == '6':
            print("Exiting the program...")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
